```python
import pandas as pd
```

```python
df=pd.read_csv("Data Model - Pizza Sales.csv")
df
```

```python
#To transform column names in upper case (for lower case replace upper with lower in df.column)

df.columns=[x.upper()for x in df.columns]
df
```

```python
df.columns=[x.lower()for x in df.columns]
df
```

```python
#Replacing value

df.columns=df.columns.str.replace("pizza_ingredients","ingredients")
df
```

```python
#Renaming a value (same as replacing)

df.rename(columns={"pizza_category":"category"},inplace="True")
df
```

```python
#To update a row 
df.loc[1,["quantity"]]=[2]
df
```

```python
#filters 

filt=(df["pizza_name"]=="The Hawaiian Pizza")
df[filt]
```
